```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("rainfall in india 1901-2015.csv")

# Step 2: Check columns (run once if unsure)
# print(df.columns)

# Step 3: Select variables
x = df['JUN']       # monsoon start
y = df['ANNUAL']    # total rainfall

# Step 4: Plot scatter
plt.figure()
plt.scatter(x, y)

plt.title("Scatter Plot: June Rainfall vs Annual Rainfall")
plt.xlabel("June Rainfall")
plt.ylabel("Annual Rainfall")

plt.show()
```
